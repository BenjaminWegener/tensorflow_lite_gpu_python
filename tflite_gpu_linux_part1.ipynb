```python
# clone tensorflow sources
!wget https://github.com/BenjaminWegener/tensorflow_lite_gpu_python/archive/refs/heads/master.zip
!unzip master.zip
```

```python
# install bazel
!wget https://releases.bazel.build/5.1.1/release/bazel-5.1.1-linux-x86_64
!chmod +x bazel-5.1.1-linux-x86_64
!sudo cp bazel-5.1.1-linux-x86_64 /usr/local/bin/bazel
```

```python
# build dependencies
%cd tensorflow_lite_gpu_python-master
!mkdir external
%cd external
!cmake ../tensorflow/lite -DCMAKE_FIND_DEBUG_MODE=1
```

```python
# build GPU delegate 
!ls
%cd ../
!ls
!bazel clean
!bazel build --local_ram_resources=HOST_RAM*0.75 --local_cpu_resources=HOST_CPUS*0.75 --config=native_arch_linux -c opt --copt "-DEGL_NO_X11" --copt="-DMESA_EGL_NO_X11_HEADERS" --copt "-DTFLITE_GPU_BINARY_RELEASE" tensorflow/lite/delegates/gpu:libtensorflowlite_gpu_delegate.so
```
